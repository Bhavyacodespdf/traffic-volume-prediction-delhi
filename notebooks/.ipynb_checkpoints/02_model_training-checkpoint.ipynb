```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
df = pd.read_csv("../data/processed/delhi_traffic_filtered.csv")
df.head()
```

```python
df["is_rush_hour"]=(
    ((df["hour"]>=8) & (df["hour"]<=11)) |
    ((df["hour"]>=17) & (df["hour"]<=21))
).astype(int)

df["is_weekend"]=(df["day"]>=5).astype(int)

df[["hour", "day", "is_rush_hour", "is_weekend"]].head(10)
```

```python
## Baseline Model: Random Forest Regressor

# Random Forest is used as a baseline ensemble model due to its robustness
# and ability to handle non-linear relationships.
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# X and y
x = df.drop(columns=[
    "traffic_volume",
    "speed",
    "free_flow"
])
y = df["speed"]

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

speed_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

speed_model.fit(x_train, y_train)
```

```python
## Model Evaluation and Comparison
```

```python
y_pred=speed_model.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

print("MAE:", mae)
print("R2 score:", r2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

importances = pd.Series(
    speed_model.feature_importances_,
    index=x.columns
).sort_values(ascending=False)

importances
```

```python
## Advanced Model: XGBoost Regressor

# XGBoost is a gradient boosting model that builds trees sequentially
# to correct previous errors.
```

```python
## Improved Model: XGBoost

from xgboost import XGBRegressor

xgb_model=XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="reg:squarederror",
    random_state=42
)

xgb_model.fit(x_train,y_train)
        
```

```python
## Model Evaluation and Comparison
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred_xgb=xgb_model.predict(x_test)

print("XGBoost MAE:", mean_absolute_error(y_test,y_pred_xgb))
print("XGBoost R2:", r2_score(y_test, y_pred_xgb))
```

```python
#Hypermeters Tuning
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
rf = RandomForestRegressor(random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring="neg_mean_absolute_error",
    n_jobs=-1
)
```

```python
grid_search.fit(x_train, y_train)
```

```python
best_rf = grid_search.best_estimator_

print("Best parameters:", grid_search.best_params_)
```

```python
import joblib
import os

# Ensure models folder exists
os.makedirs("../models", exist_ok=True)

# SAVE FINAL TRAINED MODEL
joblib.dump(best_rf, "../models/final_speed_model.pkl")

print("✅ Model saved successfully")

```

```python
import joblib

model = joblib.load("../models/final_speed_model.pkl")
print(hasattr(model, "estimators_"))
print(list(model.feature_names_in_))
```

```python
y_pred_best = best_rf.predict(x_test)

print("Tuned RF MAE:", mean_absolute_error(y_test, y_pred_best))
print("Tuned RF R2:", r2_score(y_test, y_pred_best))
```
